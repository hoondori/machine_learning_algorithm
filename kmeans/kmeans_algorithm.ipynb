# Data Creation

```python
from sklearn.datasets import make_blobs

n_samples = 500000
n_features = 2
cluster_std = 1.0
n_centers = 10
random_state = 175
X, y = make_blobs(centers = n_centers, n_samples=n_samples, n_features=n_features, cluster_std=cluster_std, random_state=random_state)
len(X)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 12))
plt.subplot(221)
plt.scatter(X[:, 0], X[:, 1])
plt.title("Incorrect Number of Blobs")
```

# KMEANS (original)

```python
import numpy as np

from timeit import default_timer as timer

from sklearn.cluster import KMeans

start = timer()
n_clusters = n_centers
y_pred = KMeans(init='random', n_clusters=n_clusters, random_state=random_state).fit_predict(X)
end = timer()
print(end - start) # Time in seconds
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(221)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Incorrect Number of Blobs")
```

# KMEANS++ 

```python
start = timer()
n_clusters = n_centers
y_pred = KMeans(init='k-means++', n_clusters=n_clusters, random_state=random_state).fit_predict(X)
end = timer()
print(end - start) # Time in seconds
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(221)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Incorrect Number of Blobs")
```

# libKMCUDA

* https://github.com/src-d/kmcuda

```python
import numpy, os
from matplotlib import pyplot
from libKMCUDA import kmeans_cuda

# convert to numpy
import numpy as np
X_npy = np.array(X)
X_npy = X_npy.astype(np.float16)

# you need libKMCUDA.so in your visible path
start = timer()
centroids, assignments = kmeans_cuda(X_npy, n_centers, verbosity=2, seed=3, device=1)
end = timer()
print(end - start) # Time in seconds
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(221)
plt.scatter(X[:, 0], X[:, 1], c=assignments)
plt.scatter(centroids[:, 0], centroids[:, 1], c="white", s=300)
```
